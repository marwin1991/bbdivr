```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "../read-write-time-results.csv"
df = pd.read_csv(path)

print(df)

ax = df.plot(x='bytes', title='Read/Wirte time to Hyperledger Fabric Blockchain network', marker='D'); 
ax.set_xlabel("message length [bytes]")
ax.set_ylabel("time [s]")
plt.legend(loc='best')

plt.savefig(path.replace('csv','png'))
```
