```python

imagesNames = ["postgres:9.6.20", "postgres:9.6.17","postgres:9.6.18", "postgres:9.6.17", "postgres:9.6.16", "postgres:9.6.15", "postgres:9.6.14","postgres:9.6.13", "postgres:9.6.12","postgres:9.6.11", "postgres:9.6.10"]
```

```python
import requests

for imageName in imagesNames:
    print(imageName)
    response = requests.get("http://localhost:8080/sum/" + imageName)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from textwrap import wrap

sum_dict = {}
df_dict = {}
sum_sev_dict = {}
for imageName in imagesNames:
    print(imageName)
    fileName = imageName.replace("/","-").replace(":","-") + "-results.csv"
    df = pd.read_csv("../../results/" + fileName)
    print(df)

    ax = df.plot(kind="bar",x='sev_name', title=imageName + ' - vulnerabilities'); 
    ax.set_xlabel("severity level")
    ax.set_ylabel("number of vulnerabilities")
    plt.legend(loc='best')

    plt.savefig("../../results/" + fileName.replace("-results.csv",".png"))
    all_sum = df.sum().get("sum")
    print("\n\nSum of all: " + str(all_sum) + "\n\n")
    sum_dict[imageName] = all_sum
    df_dict[imageName] = df
    sum_sev_dict[imageName] = df[['sev_name','sum']].set_index('sev_name').to_dict().get("sum")
    
print(sum_sev_dict)
sum_df = pd.DataFrame(sum_sev_dict);
print(sum_df.T)
labels = list(sum_df.T.index)
labels = ['\n'.join(wrap(l, 12)) for l in labels]
print(labels)
ax = sum_df.T.plot(kind='bar', figsize=(14,4))
ax.set_xlabel("image name")
ax.set_ylabel("number of vulnerabilities")
plt.xticks(range(0,len(labels)), labels, rotation='horizontal')

pref = "postgres"
sum_df.T.to_csv(f"../../results/{pref}-images-summary.csv")
plt.savefig(f"../../results/{pref}-images-summary.png", bbox_inches="tight", pad_inches=0.5)
```
